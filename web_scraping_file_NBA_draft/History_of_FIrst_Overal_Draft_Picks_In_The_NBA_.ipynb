```python
#remove special characters from the name column
draft_df[['Player Name', 'Junk']] = draft_df[
    'Player'].str.split('^', 1, expand=True)

#remove special characters from the name column
draft_df[['Player Name_2', 'More Junk']] = draft_df[
    'Player Name'].str.split('[', 1, expand=True)
#drop irrelevant columns containing special characters
draft_df = draft_df.drop(['Player',
                          'Player Name', 'Junk', 'More Junk'], axis=1)

#organise column positions
draft_df = draft_df.reindex(columns=['Draft', 'Selected by','Player Name_2',
                                     'Nationality', 'Position',
                                     'College/Overseas','PPG', 'RPG', 'APG'])

#rename player column
draft_df = draft_df.rename(columns={'Player Name_2': 'Player'})
```
